Есть два набора точек - *x_hw* и *y_hw*. В рамках домашней работы нужно
* визуализируйте набор точек
* постройте коэффициенты регрессию $w_0, w_1$ по шагам, как в уроке
* посчитайте предсказание в виде $\hat{y} = w_0 + w_1x$ и визуализируйте его вместе с точками *x_hw* и *y_hw*

```python
x_hw = [50, 60, 70, 100]
y_hw = [10, 15, 40, 45]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(x_hw, y_hw)
```

```python
import numpy as np
from numpy.linalg import inv
X=np.array([[1,50],[1,60],[1,70],[1,100]])
Y=np.array(y_hw)
X_T_X = (X.T).dot(X)
X_T_X_inverted = inv(X_T_X)
w = X_T_X_inverted.dot(X.T).dot(Y)
w
```

```python
%matplotlib inline
margin = 10
X_min = 20
X_max = X.max()+margin
X_max
X_support = np.linspace(X_min, X_max, num=100)
Y_model = w[0] + w[1]*X_support
plt.xlim(X_min, X_max)
plt.ylim(0, Y.max() + margin)
plt.scatter(x_hw, y_hw, 40, 'g', 'o', alpha=0.8)
plt.plot(X_support, Y_model)
plt.show()
```
